```python
from pycocotools.coco import COCO
from PIL import Image  # For loading images
from tqdm import tqdm
```

```python
import os
import sys

current_dir = os.path.dirname(os.path.abspath(os.getcwd()))
src_dir = os.path.join(current_dir, "src")
sys.path.append(current_dir)
sys.path.append(src_dir)


from src.utils.filtering_utils import FiltersUtils # noqa
```

```python
# Data can be downloaded from
# https://www.kaggle.com/datasets/awsaf49/coco-2017-dataset/data


data_folder = "../data"
train_folder = f"{data_folder}/train2017"
captions_json = f"{data_folder}/annotations/captions_train2017.json"
coco = COCO(captions_json)
```

```python
all_image_files = [
    f for f in os.listdir(train_folder) if os.path.isfile(os.path.join(train_folder, f))
]

filtered_image_files = []
for file in tqdm(all_image_files, desc="Filtering images"):
    img_path = os.path.join(train_folder, file)
    cause = "cause: channel_difference_test"
    should_filter = False
    if FiltersUtils.channel_difference_test(img_path):
        should_filter = True

    if FiltersUtils.caption_keywords_test(img_path, coco):
        cause = "cause: caption_keywords_test"
        should_filter = True

    if should_filter:
        filtered_image_files.append((img_path, cause))
```

Keep in mind that channel_difference_test is performed first so it might filter out more images.

```python
from typing import List
from IPython.display import display
from pathlib import Path

print(
    f"Filtered {len(filtered_image_files)} images out of {len(all_image_files)} total images."
)


def get_captions(img_path: str) -> List[str]:
    """
    Extracts captions from the COCO dataset for a given image path.
    Args:
        img_path (str): Path to the image file.
    Returns:
        List[str]: List of strings representing captions for the image.
    """
    img_id = int(
        Path(img_path).stem.split("_")[-1]
    )  # e.g. train2017/000000391895.jpg → 391895
    anns = coco.imgToAnns[img_id]
    return [a["caption"] for a in anns]


for img_path, cause in filtered_image_files[500:515]:
    print(f"Filtered {img_path} due to {cause}")
    display(Image(filename=img_path))
    print("Captions:")
    captions = get_captions(img_path)
    for caption in captions:
        print(f"- {caption}")
    print("\n")
    print("========================================")
```
